## Como tudo começou...

###### Meus pais decidiram presentear seu único filho (até o momento) com um irmãozinho, e no dia 04 de novembro do ano de 1991, na cidade de Entre Rios de Minas, mais precisamente nesse local tudo começou...

```python


image_filename = "map_hcc.png"
encoded_image = base64.b64encode(open(image_filename, 'rb').read())
img = 'data:image/png;base64,{}'.format(encoded_image.decode())

layout = go.Layout(
    images=[go.layout.Image(
        source=img, 
        sizex=1, sizey=1, 
        yanchor="bottom"
    )],
    template="plotly_white",
    title = 'Local de Nascimento: Hospital Cassiano Campolina - Entre Rios de Minas',
    xaxis = go.XAxis(
        showticklabels=False),
    yaxis = go.YAxis(
        showticklabels=False),
  
)
    
fig = go.Figure(layout=layout)
fig.show()
```

```python
hobbys_kids = {
    'Atividade': ['Videogame','Futebol', 'Natação', 'Leitura', 'Escola'],
    'Tempo_Gasto' : [25,15,20,15,35],
    'Habilidade' : [7,3,8,7,9]
}
fig = px.bar(hobbys_kids, x =hobbys_kids['Atividade'] , y =hobbys_kids['Tempo_Gasto'],labels={'x':'Atividade', 'y' : 'Tempo gasto em %'})
fig.update_traces(marker_color='rgb(128,128,128)', marker_line_color='rgb(105,105,105)',
                  marker_line_width=1.5)
fig.update_layout(title_text='Tempo gasto em meus principais hobbies')
```

```python
fig = go.Figure(data=go.Scattergeo(
    
    lat = [-19.6248, 40.4781,48.8589,51.5076,52.3547,41.3925],
    lon = [-43.9663, -3.7034,2.3469,-0.1276,4.9040,2.1404],
    #name= ['Belo Horizonte','Madrid','Paris','Londres','Amsterdam'],
    mode = 'lines',
    line = dict(width = 4, color = 'red'),
))

fig.update_layout(
    title_text = 'Minha primeira viagem iternacional',
    showlegend = False,
    geo = dict(
        resolution = 50,
        showland = True,
        showlakes = True,
        landcolor = 'rgb(204, 204, 204)',
        countrycolor = 'rgb(204, 204, 204)',
        lakecolor = 'rgb(255, 255, 255)',
        projection_type = "equirectangular",
        coastlinewidth = 2,
        lataxis = dict(
            range = [-20, 60],
            showgrid = True,
            dtick = 10
        ),
        lonaxis = dict(
            range = [-100, 20],
            showgrid = True,
            dtick = 20
        ),
    )
)

fig.show()
```
